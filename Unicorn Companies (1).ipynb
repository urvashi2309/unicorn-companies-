```python
v.dropna(inplace=True)
```

```python
v=v.reset_index()
```

```python
v.head(10)
```

```python
for i in range(len(v["Valuation"])):
    k=str(v["Valuation"][i])
    if "M" in k:
        k=k.replace("M","")
        k=k.replace("$","")
        v.loc[i,"Valuation"]=int(k)*1000000
    else:
        k=k.replace("B","")
        k=k.replace("$","")
        v.loc[i,"Valuation"]=int(k)*1000000000
```

```python
for i in  range(len(v["Valuation"])):
    if v.loc[i,"Funding"]=="Unknown":
        v.loc[i,"Funding"]=v.loc[i,"Valuation"]
```

```python
for i in range(len(v["Valuation"])):
    k=v["Funding"][i]
    if "M" in str(k):
        k=k.replace("M","")
        k=k.replace("$","")
        v.loc[i,"Funding"]=int(k)*1000000
    elif "B" in str(k):
        k=k.replace("B","")
        k=k.replace("$","")
        v.loc[i,"Funding"]=int(k)*1000000000
```
